# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

### Permutation Test

#### 1. Setup an appropriate hypothesis test.
* **Null Hypothesis**: There is a negative relationship between number of discharges and excess of re-admission. 
* **Alternate hypothesis**: there is no negative relationship between the number of discharges and excess of re-admission.
* **Test Statistic**: Pearson correlation coefficient, ρ, between number of discharges & excess rate of readmissions.

#### 2. Compute and report the observed significance value (or p-value).

```python
# define function for calculate pearson correlation, ρ
def pearson_r(x, y):
    """Compute Pearson correlation coefficient between two arrays."""
    corr_mat = np.corrcoef(x, y)
    return corr_mat[0,1]

r_obs = pearson_r(x, y)
print(r_obs)
```

```python
# perform permutation test
perm_replicates = np.empty(10000)

for i in range(10000):
    x_permuted = np.random.permutation(x)
    perm_replicates[i] = pearson_r(x_permuted, y)

p = np.sum(perm_replicates <= r_obs) / len(perm_replicates)
print('p-value: ', p)
```

#### 3. Report statistical significance for $\alpha$ = .01. 

```python
alpha = 0.01
if p < alpha:
    print('p-value is less than 0.01.')
else:
    print('p-value is equal or more than than 0.01.')
```

#### Summary:
1. Pearson correlation coefficient is -0.09. As person correlation coefficient lies between -1 to 1, -0.09 means when number of discharges increases, excess of re-admission decreases, but the correlation is pretty weak.
2. As p-value is less than $\alpha$, we can reject the null hypothesis, which means that there is inconclusive evidence that there is a negative relationship between number of discharges and excess of re-admission.

> * Construct another plot that conveys the same information in a more direct manner.

Let's split the dataset into 20-evenly sized groups in sorted order from least to most number of discharges.

```python
pct = np.percentile(sorted(clean_hospital_read_df['Number of Discharges']), range(5, 105, 5))
print(pct)

def group_by_n_discharges(x):
    group = 0
    lastP = 0
    for i, p in enumerate(pct):
        if x > lastP and x <= p:
            group = i + 1
            break
        lastP = p
    return group
        
clean_hospital_read_df['Number of Discharges Group'] = clean_hospital_read_df['Number of Discharges'].apply(group_by_n_discharges)
```

```python
def get_sub_plot(width = 20, height = 10, label_font_size = 14, tick_font_size = 13):
    fig, ax = plt.subplots()
    fig.set_size_inches(width, height)
    ax.tick_params(labelsize = tick_font_size)
    ax.set_xlabel(ax.get_xlabel(), fontsize = label_font_size)
    ax.set_ylabel(ax.get_ylabel(), fontsize = label_font_size)
    return ax

ax = get_sub_plot()
sns.violinplot(x = 'Number of Discharges Group', y = 'Excess Readmission Ratio', data = clean_hospital_read_df, ax = ax)
```

```python
clean_hospital_read_df_grouped = clean_hospital_read_df.groupby('Number of Discharges Group').aggregate(np.mean).reset_index()
clean_hospital_read_df_grouped[['Number of Discharges Group', 'Excess Readmission Ratio']]
```

```python
ax = get_sub_plot()
sns.barplot(x = 'Number of Discharges Group', y = 'Excess Readmission Ratio', data = clean_hospital_read_df_grouped, ax = ax)
```

Based on the boxplot above, we can see that with the data points evenly-distributed, when number of discharges increases, the range of readmissions also increases. If we further plot the mean for each group in a barplot we can see that the mean of readmission ratio apparently decrease at first two to three groups.
